## Data Processing

Primero, importamos las liberias que vamos a usar dentro de este jupyter notebook

```python
import pandas as pd
import json
import datetime
```

Ahora, abrimos el archivo con los datos

```python
unclean_data_df  = pd.read_json('farmers-protest-tweets-2021-03-5.json', lines = True)
unclean_data_df.head()
```

Queremos limpiar un poco el dataframe para eliminar valores duplicados, en este caso no es necesario normalizar los datos dado que nos interesa el número exacto de tweets.

No eliminamos los nulos dado que algúnos usuarios simplemente tienen valores nulos en retweetedTweet, quotedTweet y mentionedUsers. 

Como podemos ver no hay datos duplicados

```python
# Primero vemos si tenemos duplicados  
clean_data_df = unclean_data_df.loc[unclean_data_df.astype(str).drop_duplicates().index]
clean_data_df.head()
```

Como podemos ver la columna 'user' esta compuesta de un dictionario username y displayname, por lo tanto vamos a crear dos columnas nuevas llamadas 'username' y 'displayname' y borramos 'user'

```python
username = []
displayname = []

for i in range(len(clean_data_df)):
    username.append(clean_data_df['user'][i]['username'])
    displayname.append(clean_data_df['user'][i]['displayname'])

clean_data_df['username'] = username
clean_data_df['displayname'] = displayname

clean_data_df.head()
```

```python
print(f"La forma solía ser {unclean_data_df.shape} y ahora es {clean_data_df.shape}")
```

## Pregunta 1

```python
def top_10_tweets_retweeted(clean_data_df):
    display(clean_data_df.nlargest(10, 'retweetCount'))
```

## Pregunta 2

```python
def top_10_users(clean_data_df):
    display(pd.DataFrame(clean_data_df.groupby(['username']).size(), columns = ['emittedTweets']).nlargest(10, 'emittedTweets'))
```

## Pregunta 3

```python
def top_10_days(clean_data_df):
    clean_data_df['Date'] = [datetime.datetime.date(d) for d in clean_data_df['date']] 
    display(pd.DataFrame(clean_data_df.groupby(['Date']).size(), columns = ['emittedTweetsDay']).nlargest(10, 'emittedTweetsDay'))
```

## Pregunta 4

```python
def getListHashtags(content):
    
    content = content.str.findall(r'#.*?(?=\s|$)') # [3]

    all_hashtags_content = []
    for i in range(len(content)):
        for j in range(len(content[i])):
            all_hashtags_content.append(content[i][j])
    
    return all_hashtags_content
```

Las listas all_hashtags_content y all_hashtags_rendered, no son iguales. Por lo tanto, vamos a sumarlas

```python
def top_10_hashtags(clean_data_df):
    all_hashtags_content = getListHashtags(clean_data_df['content'])
    all_hashtags_rendered = getListHashtags(clean_data_df['renderedContent'])
    
    print(f"Las listas all_hashtags_content y all_hashtags_rendered son iguales: {all_hashtags_content == all_hashtags_rendered}")
    all_hashtags = all_hashtags_rendered + all_hashtags_content

    all_hashtags_df = pd.DataFrame(all_hashtags, columns = ["hashtag"])
    display(pd.DataFrame(all_hashtags_df.groupby(['hashtag']).size(), columns = ['countHashtag']).nlargest(10, 'countHashtag'))
    
```

## Main 

```python
def main(question_number):
    if question_number == 1:
        return top_10_tweets_retweeted(clean_data_df)
        
    elif question_number == 2:
        return top_10_users(clean_data_df)
        
    elif question_number == 3:
        return top_10_days(clean_data_df)
        
    elif question_number == 4:
        return top_10_hashtags(clean_data_df)
```

```python
if __name__ == "__main__":
    question_number = int(input("Agrege el número de pregunta que desea ejecutar : "))
    main(question_number)
```
